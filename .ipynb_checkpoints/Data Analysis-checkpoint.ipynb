### Total number of tweets

```python
df_tweet['hashtags'] = df_tweet['hashtags'].replace(np.nan, "['None']", regex=True)
df_tweet['hashtags'] = df_tweet['hashtags'].apply(lambda x: x.replace('\\N',''))
df_tweet['hashtags_count'] = df_tweet['hashtags'].apply(lambda x: len(x.split(',')))
```

```python
df_tweet['hashtags_count'] = df_tweet['hashtags'].apply(lambda x: len(x.split(',')))
```

```python
print(df_tweet["hashtags"].sum())
```

```python
all_tweet_list = []

# Itertuples is much faster than iterrows
for each_row in df_tweet.itertuples():
    if not str(each_row.tweet).lower() == "nan":
        each_tweet = str(each_row.tweet)
        each_tweet = each_tweet.strip('@').replace("'","")
        all_tweet_list += each_tweet.split(",")
        
print("Total number of tweets",len(all_tweet_list))
```

```python
all_hashtags_list = []

# Itertuples is much faster than iterrows
for each_row in df_tweet.itertuples():
    if not str(each_row.hashtags).lower() == "nan":
        each_hashtag = str(each_row.hashtags)
        each_hashtag = each_hashtag.strip('[]').replace("'","")
        all_hashtags_list += each_hashtag.split(",")
        
print("Total number of hashtags",len(all_hashtags_list))
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.corpus import stopwords
```

```python
stop_words = set(STOPWORDS)
stop_words.add("xa0")
stop_words.add("https")
stop_words.add("tweet")
stop_words.add("twitter")
stop_words.add("pic")
stop_words.add("i")
stop_words.add("com")

def create_wordcloud(tweets, bg_color):
    wordcloud = WordCloud(
        background_color=bg_color,
        width=1000,
        height=500,
        contour_color='steelblue',
        stopwords=stop_words).generate(str(df_tweet.tweet))
    return wordcloud
```

```python
plt.rcParams["figure.figsize"] = 30, 30
wordcloud = create_wordcloud(set(df_tweet), "black")
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
```
